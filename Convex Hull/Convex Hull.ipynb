# Convex Hull 101

## An Example

```python
import numpy as np
from scipy.spatial import ConvexHull
import matplotlib.pyplot as plt
% matplotlib notebook
```

```python
numberOfData=40
pointCloud2D=0.1 + 0.8*np.random.rand(numberOfData, 2)
pointCloud2D
```

```python
ch=ConvexHull(points=pointCloud2D)
ch
```

```python
hullIndices=ch.vertices
hullIndicesEnclosed=np.concatenate([hullIndices,np.array([hullIndices[0]])])
```

```python
plt.figure(1)
plt.scatter(x=pointCloud2D[:,0],y=pointCloud2D[:,1],label="data")
plt.plot(pointCloud2D[hullIndicesEnclosed,0],pointCloud2D[hullIndicesEnclosed,1],label="edges of the convex hull")
plt.title("Convex Hull of dataset")
plt.legend()
plt.show()
```
